```python
user_input = input("Please put three numbers with commas between each one: ")

numbers = [int(num.strip()) for num in user_input.split(',') if num.strip()]

if numbers:
    max_num = max(numbers)
    print(f"The biggest number you entered is: {max_num}")
else:
    print("No numbers were entered.")
```
